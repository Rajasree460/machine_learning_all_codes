```python
#fuzzy uni inter comp

import numpy as np
import matplotlib.pyplot as plt

#define fuzzy set A & B
np.random.seed(42)
x=np.arange(100,201)  #X==A==B

#define ms vals
ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#uni
fuzzy_uni=np.maximum(ms_A,ms_B)  #ms_uni
#inter
fuzzy_inter=np.minimum(ms_A,ms_B)  #ms_inter
#comp
comp_A=np.subtract(1,ms_A)  #ms_comp
comp_B=np.subtract(1,ms_B)

#plotting
plt.figure(figsize=(12,8))
plt.plot(x,ms_A,label="set A",marker='o')
plt.plot(x,ms_B,label="set B",marker='o')
plt.plot(x,fuzzy_uni,label="fuzzy union",marker='o')
plt.plot(x,fuzzy_inter,label="fuzzy intersection",marker='o')
plt.plot(x,comp_A,label="fuzzy complement A",marker='o')
plt.plot(x,comp_B,label="fuzzy complement B",marker='o')
plt.xlabel("vals")
plt.ylabel("ms vals/grades")
plt.title("fuzzy un inter comp")
plt.legend()
plt.grid(True)
plt.show()
```
