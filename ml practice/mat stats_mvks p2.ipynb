```python
#mat stats mvks

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

#100 x 10 matrix
np.random.seed(42)
mat=np.random.rand(100,10)

#convert mat into df
data=pd.DataFrame(mat)

#calc mvks
mean=data.mean()
varience=data.var()
kurtosis=stats.kurtosis(data)
skewness=stats.skew(data)

#print
print("mean:\n",mean)
print("\nvarience:\n",varience)
print("\nkurtosis:\n",kurtosis)
print("\nskewness:\n",skewness)
print("\noverall mean:",np.mean(mean))
print("\noverall varience:",np.var(varience))
print("\noverall kurtosis:",stats.kurtosis(kurtosis))
print("\nskewness:",stats.skew(skewness))

#plot
plt.figure(figsize=(12,8))
plt.plot(mean,marker='o',label="mean")
plt.plot(varience,marker='o',label="varience")
plt.plot(kurtosis,marker='s',label="kurtosis")
plt.plot(skewness,marker='x',label="skewness")
plt.title("mvks plot")
plt.legend()
plt.grid(True)
plt.show()


#statistical properties: histogram
fig,axs=plt.subplots(2,2,figsize=(12,8))

#mean
axs[0,0].hist(mean,bins=10,alpha=0.7)
axs[0,0].set_title("plotting mean")
axs[0,0].set_xlabel("mean")
axs[0,0].set_ylabel("freq")

#var
axs[0,1].hist(varience,bins=10,alpha=0.7)
axs[0,1].set_title("plotting var")
axs[0,1].set_xlabel("var")
axs[0,1].set_ylabel("freq")

#kurtosis
axs[1,0].hist(kurtosis,bins=10,alpha=0.7)
axs[1,0].set_title("plotting kurtosis")
axs[1,0].set_xlabel("kurtosis")
axs[1,0].set_ylabel("freq")

#skew
axs[1,1].hist(skewness,bins=10,alpha=0.7)
axs[1,1].set_title("plotting skewness")
axs[1,1].set_xlabel("skewness")
axs[1,1].set_ylabel("freq")

plt.tight_layout()
plt.show()
```
