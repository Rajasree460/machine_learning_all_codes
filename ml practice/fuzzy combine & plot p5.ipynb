```python
#fuzzy combine & plot

import numpy as np
import matplotlib.pyplot as plt

#define A & B
np.random.seed(42)
num_eles=10
params_A=np.random.uniform(low=0,high=10,size=(num_eles,3))
params_B=np.random.uniform(low=0,high=10,size=(num_eles,3))

#define tri MF
def tri(x,a,b,c):
    return np.where(x<=a,0,
                   np.where(x<b,(x-a)/(b-a),
                           np.where(x<c,(c-x)/(c-b),
                                   0)))

#define x(an array of 100 evenly spaced vals starting from 0 ending at 11)
x=np.linspace(0,11,100)

#define ms vals
ms_A_list=[]
ms_B_list=[]
ms_union_list=[]

for (a1,b1,c1),(a2,b2,c2) in zip(params_A,params_B):
    ms_A=tri(x,a1,b1,c1)
    ms_B=tri(x,a2,b2,c2)
    ms_union=np.maximum(ms_A,ms_B)

    ms_A_list.append(ms_A)
    ms_B_list.append(ms_B)
    ms_union_list.append(ms_union)

#plotting
#A
plt.figure(figsize=(12,8))
for i,ms_A in enumerate(ms_A_list):
    plt.plot(x,ms_A,label=f'A{i+1}')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.title("plotting fuzzy set A")
plt.legend()
plt.grid(True)
plt.show()

#B
plt.figure(figsize=(12,8))
for i,ms_B in enumerate(ms_B_list):
    plt.plot(x,ms_B,label=f'B{i+1}')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.title("plotting fuzzy set B")
plt.legend()
plt.grid(True)
plt.show()

#A U B
plt.figure(figsize=(12,8))
for i,ms_union in enumerate(ms_union_list):
    plt.plot(x,ms_union,label=f'A{i+1} U B{i+1}')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.title("plotting fuzzy set A U B")
plt.legend()
plt.grid(True)
plt.show()
```
