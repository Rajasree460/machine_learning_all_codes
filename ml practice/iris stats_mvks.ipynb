```python
#iris stats_mvks

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

#load the data
data=pd.read_csv("C:/Users/HP/OneDrive/Desktop/ml 7th sem codes/datasets/iris.csv")
data
```

```python
data=data.drop('variety',axis=1)
data
```

```python
#calculate
mean=data.mean()
var=data.var()
kurtosis=stats.kurtosis(data)
skewness=stats.skew(data)

#print
print("mean:\n",mean)
print("\nvarience:\n",var)
print("\nkurtosis:\n",kurtosis)
print("\nskewness:\n",skewness)
print("\noverall mean:",np.mean(mean))
print("\noverall varience:",np.var(var))
print("\noverall kurtosis:",stats.kurtosis(kurtosis))
print("\noverall skewness:",stats.skew(skewness))
```

```python
#plot
plt.figure(figsize=(12,8))
plt.plot(mean,marker='o',label="mean")
plt.plot(var,marker='o',label="varience")
plt.plot(kurtosis,marker='s',label="kurtosis")
plt.plot(skewness,marker='x',label="skewness")
plt.title("mvks plot")
plt.legend()
plt.grid(True)
plt.show()
```
