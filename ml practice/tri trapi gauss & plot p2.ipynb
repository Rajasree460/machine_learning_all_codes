```python
# tri trapi gauss & plot

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#define set A & Ms vals
np.random.seed(42)
A=np.arange(1,11)
ms_A=np.random.rand(len(A))

#define x(an array with 1000 evenly spaced vals starts from 0 & ends with 11)
x=np.linspace(0,11,1000)

#define tri trapi gauss MF
def tri(x,a,b,c):
    return np.maximum(np.minimum((x-a)/(b-a),(c-x)/(c-b)),0)

def trapi(x,a,b,c,d):
    return np.maximum(np.minimum(np.minimum((x-a)/(b-a),1),(d-x)/(d-c)),0)

def gauss(x,mu,sigma):
    return stats.norm.pdf(x,mu,sigma)

#main func
tri_params=tuple(map(float,input("enter the parameters for tri MF(x,a,b,c) seperated by spaces:").split()))
ms_tri=tri(x,*tri_params)

trapi_params=tuple(map(float,input("enter the parameters for trapi MF(x,a,b,c,d) seperated by spaces:").split()))
ms_trapi=trapi(x,*trapi_params)

gauss_params=tuple(map(float,input("enter the parameters for gauss MF(x,mu,sigma) seperated by spaces:").split()))
ms_gauss=gauss(x,*gauss_params)

#fit & plot
plt.figure(figsize=(12,8))
plt.scatter(A,ms_A,color='red',zorder=5)
plt.plot(x,ms_tri,label="tri")
plt.plot(x,ms_trapi,label="trapi")
plt.plot(x,ms_gauss,label="gauss")
plt.title("plotting & fitting tri trapi & gauss MF")
plt.xlabel("A or x or vals")
plt.ylabel("member grades/vals")
plt.legend()
plt.show()
```
