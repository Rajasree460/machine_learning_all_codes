```python
# 4 prin on composite relationship

import numpy as np
import matplotlib.pyplot as plt

#define the relation R & S
np.random.seed(42)
R=np.random.rand(5,6)
S=np.random.rand(6,3)

#define com. funcs
#max_min func
def max_min_comp(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max(np.minimum(R[i,:],S[:,j]))

    return res


#max_min func
def min_max_comp(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.min(np.maximum(R[i,:],S[:,j]))

    return res

#max_prouct func
def max_product_comp(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max(R[i,:]*S[:,j])

    return res

#max_avg func
def max_avg_comp(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max((R[i,:]+S[:,j])/2)

    return res


#main func
#calc compositions
max_min_res=max_min_comp(R,S)
min_max_res=min_max_comp(R,S)
max_product_res=max_product_comp(R,S)
max_avg_res=max_avg_comp(R,S)

#print results
print("R:\n",R)
print("S:\n",S)
print("max min composition:\n",max_min_res)
print("min max composition:\n",min_max_res)
print("max product composition:\n",max_product_res)
print("max avg composition:\n",max_avg_res)



#plot results
def plot_mat(mat,title):
    plt.figure(figsize=(12,8))
    plt.imshow(mat,cmap='viridis',aspect='auto')
    plt.title(title)
    plt.colorbar()
    plt.show()
    
plot_mat(max_min_res,"max min composition")
plot_mat(min_max_res,"min max composition")
plot_mat(max_product_res,"max product composition")
plot_mat(max_avg_res,"max avg composition")
```
