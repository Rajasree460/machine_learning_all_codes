```python
#fuzzy uni inter com

import numpy as np
import matplotlib.pyplot as plt

#define set fuzzy A & B
np.random.seed(42)
x=np.arange(100,201) #A==B==x

#define membership vals of A & B
ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#define fuzzy uni
fuzzy_uni=np.maximum(ms_A,ms_B)

#define fuzzy inter
fuzzy_inter=np.minimum(ms_A,ms_B)

#define com A & B
com_A=np.subtract(1,ms_A)
com_B=np.subtract(1,ms_B)

#plotting
plt.figure(figure=(12,8))
plt.plot(x,ms_A,label="fuzzy set A",marker='o')
plt.plot(x,ms_B,label="fuzzy set B",marker='o')
plt.plot(x,fuzzy_uni,label="A U B",marker='o')
plt.plot(x,fuzzy_inter,label="A  B",marker='o')
plt.plot(x,com_A,label="complement A",marker='o')
plt.plot(x,com_B,label="complement B",marker='o')
plt.xlabel("values/eles")
plt.ylabel("membership vals/grades")
plt.title("fuzzy uni inter com plot")
plt.legend()
plt.grid(True)
plt.show()
```
