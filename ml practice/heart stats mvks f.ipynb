```python
#heart stats mvks 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

#load the ds
data=pd.read_csv("C:/Users/HP/OneDrive/Desktop/ml 7th sem codes/datasets/heart.csv")
data
```

```python
#drop the cl
data=data.drop("target",axis=1)
data
```

```python
#calc(colwise mvks)
mean=data.mean()
var=data.var()
kurto=stats.kurtosis(data)
skew=stats.skew(data)
```

```python
#print
print("mean:")
print(mean)
print("var:")
print(var)
print("kurto:")
print(kurto)
print("skew:")
print(skew)

print("\noverall mean:",np.mean(mean))
print("\noverall var:",np.var(var))
print("\noverall kurto:",stats.kurtosis(kurto))
print("\noverall skew:",stats.skew(skew))
```

```python
#plot
plt.figure(figsize=(12,8))
plt.plot(mean,label="mean",marker='o')
plt.plot(var,label="var",marker='o')
plt.plot(kurto,label="kurtosis",marker='s')
plt.plot(skew,label="skewness",marker='x')
plt.title("plotting mvks")
plt.legend()
plt.grid(True)
plt.show()
```
