```python
#fuzzy uni inter comp

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x=np.arange(100,201) #x==A==B

ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#uni
fuzzy_uni=np.maximum(ms_A,ms_B)
#inter
fuzzy_inter=np.minimum(ms_A,ms_B)
#complement
comp_A=np.subtract(1,ms_A)
comp_B=np.subtract(1,ms_B)

#plot
plt.figure(figsize=(12,8))
plt.plot(x,ms_A,label="set A",marker='o')
plt.plot(x,ms_B,label="set B",marker='o')
plt.plot(x,fuzzy_uni,label="fuzzy union",marker='o')
plt.plot(x,fuzzy_inter,label="fuzzy intersection",marker='o')
plt.plot(x,comp_A,label="complement A",marker='o')
plt.plot(x,comp_B,label="complement B",marker='o')
plt.title("plotting fuzzy uni inter comp")
plt.xlabel("eles")
plt.ylabel("m values")
plt.legend()
plt.grid(True)
plt.show()
```
