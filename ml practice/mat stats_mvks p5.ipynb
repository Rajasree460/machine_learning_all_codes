```python
#mat stats_mvks

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

#define mat
np.random.seed(42)
mat=np.random.rand(100,10)

#convert mat into df
data=pd.DataFrame(mat)

#calc
mean=data.mean()
varience=data.var()
kurtosis=stats.kurtosis(data)
skewness=stats.skew(data)

#print
print("mean:")
print(mean)
print("\nvarience:")
print(varience)
print("\nkurtosis:")
print(kurtosis)
print("skewness:")
print(skewness)

print("\noverall mean:",np.mean(mean))
print("\noverall varience:",np.var(varience))
print("\noverall kurtosis:",stats.kurtosis(kurtosis))
print("\noverall skewness:",stats.skew(skewness))

#plot
plt.figure(figsize=(12,8))
plt.plot(mean,label="mean",marker='o')
plt.plot(varience,label="varience",marker='o')
plt.plot(kurtosis,label="kurtosis",marker='o')
plt.plot(skewness,label="skewness",marker='o')
plt.title("plotting mvks")
plt.legend()
plt.grid(True)
plt.show()

#histogram
fig,axs=plt.subplots(2,2,figsize=(12,8))

axs[0,0].hist(mean,bins=10,alpha=0.7,color='red')
axs[0,0].set_title("histogram of mean")
axs[0,0].set_xlabel("mean vals")
axs[0,0].set_ylabel("freq")

axs[0,1].hist(varience,bins=10,alpha=0.7,color='green')
axs[0,1].set_title("histogram of varience")
axs[0,1].set_xlabel("varience vals")
axs[0,1].set_ylabel("freq")

axs[1,0].hist(kurtosis,bins=10,alpha=0.7,color='yellow')
axs[1,0].set_title("histogram of kurtosis")
axs[1,0].set_xlabel("kurtosis vals")
axs[1,0].set_ylabel("freq")

axs[1,1].hist(skewness,bins=10,alpha=0.7,color='blue')
axs[1,1].set_title("histogram of skewness")
axs[1,1].set_xlabel("skewness vals")
axs[1,1].set_ylabel("freq")

plt.tight_layout()
plt.show()
```
