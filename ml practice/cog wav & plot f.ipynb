```python
#cog wav & plot 

import numpy as np
import matplotlib.pyplot as plt

#define A & B
np.random.seed(42)
x=np.random.randint(1,101,20)  #x==A==B

#define ms vals of A & B
ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#calc cog & wav
cog_A=sum(x*ms_A)/sum(ms_A)
cog_B=sum(x*ms_B)/sum(ms_B)
wav_A=sum(x*ms_A)/sum(ms_A)
wav_B=sum(x*ms_B)/sum(ms_B)

#print 
print("fuzzy set A & B:",x)
print("\nms vals of A:",ms_A)
print("\nms vals of B:",ms_B)
print("\ncog of A:",cog_A)
print("\ncog of B:",cog_B)
print("\nwav of A:",wav_A)
print("\ncog of B:",wav_B)

#plotting
#set A
plt.figure(figsize=(12,8))
plt.scatter(x,ms_A,label="ms val",color='orange')
plt.plot(x,ms_A,color='orange')
plt.axvline(cog_A,linestyle="--",color='yellow',label=f'cog:{cog_A:.2f}')
plt.axvline(wav_A,linestyle=":",color='blue',label=f'wav:{wav_A:.2f}')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.title("fuzzy set A")
plt.legend()
plt.show()

#set B
plt.figure(figsize=(12,8))
plt.scatter(x,ms_B,label="ms val",color='orange')
plt.plot(x,ms_A,color='orange')
plt.axvline(cog_B,linestyle="--",color='yellow',label=f'cog:{cog_B:.2f}')
plt.axvline(wav_B,linestyle=":",color='blue',label=f'wav:{wav_B:.2f}')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.title("fuzzy set B")
plt.legend()
plt.show()
```
