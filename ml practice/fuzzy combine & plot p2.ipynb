```python
#fuzzy combine & plot 

import numpy as np
import matplotlib.pyplot as plt

#define set A & B
np.random.seed(42)
num_eles=10 #no. of eles in each fuzzy sets
param_A=np.random.uniform(low=0,high=10,size=(num_eles,3))
param_B=np.random.uniform(low=0,high=10,size=(num_eles,3))

#define x
x=np.linspace(0,10,100)

#define tri MF
def tri(x,a,b,c):
    return np.where(x<=a,0,
                   np.where(x<b,(x-a)/(b-a),
                           np.where(x<c,(c-x)/(c-b),0)))

#ms vals
mv_A_list=[]
mv_B_list=[]
mv_union_list=[]

for (a1,b1,c1),(a2,b2,c2) in zip(param_A,param_B):
    mv_A=tri(x,a1,b1,c1)
    mv_B=tri(x,a2,b2,c2)
    mv_union=np.maximum(mv_A,mv_B)

    mv_A_list.append(mv_A)
    mv_B_list.append(mv_B)
    mv_union_list.append(mv_union)


#plot
plt.figure(figsize=(12,8))
for i,mv in enumerate(mv_A_list):
    plt.plot(x,mv,label=f'fuzzy set A{i+1}')
plt.title("fuzzy set A")
plt.xlabel("x or A or vals")
plt.ylabel("ms vals/grades")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12,8))
for i,mv in enumerate(mv_B_list):
    plt.plot(x,mv,label=f'fuzzy set B{i+1}')
plt.title("fuzzy set B")
plt.xlabel("x or B or vals")
plt.ylabel("ms vals/grades")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12,8))
for i,mv in enumerate(mv_union_list):
    plt.plot(x,mv,label=f'fuzzy set A{i+1} U B{i+1}')
plt.title("fuzzy set A U B")
plt.xlabel("x or A or B or vals")
plt.ylabel("ms vals/grades")
plt.legend()
plt.grid(True)
plt.show()
```
