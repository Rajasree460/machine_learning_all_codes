```python
#mat stats mvks

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

#define the mat
np.random.seed(42)
mat=np.random.rand(100,10)

#convert mat into df
data=pd.DataFrame(mat)

#calc(colwise mvks)
mean=data.mean()
var=data.var()
kurto=stats.kurtosis(data)
skew=stats.skew(data)

#print
print("mean:")
print(mean)
print("var:")
print(var)
print("kurto:")
print(kurto)
print("skew:")
print(skew)

print("\noverall mean:",np.mean(mean))
print("\noverall var:",np.var(var))
print("\noverall kurto:",stats.kurtosis(kurto))
print("\noverall skew:",stats.skew(skew))

#plot
plt.figure(figsize=(12,8))
plt.plot(mean,label="mean",marker='o')
plt.plot(var,label="var",marker='o')
plt.plot(kurto,label="kurtosis",marker='o')
plt.plot(skew,label="skewness",marker='o')
plt.title("plotting mvks")
plt.legend()
plt.grid(True)
plt.show()

#stats prop(histogram)
fig,axs=plt.subplots(2,2,figsize=(12,8))

#mean
axs[0,0].hist(mean,bins=10,alpha=0.7,color='red')
axs[0,0].set_xlabel("mean")
axs[0,0].set_ylabel("freq")
axs[0,0].set_title("hist of mean")

#var
axs[0,1].hist(var,bins=10,alpha=0.7,color='yellow')
axs[0,1].set_xlabel("var")
axs[0,1].set_ylabel("freq")
axs[0,1].set_title("hist of var")

#kurto
axs[1,0].hist(kurto,bins=10,alpha=0.7,color='blue')
axs[1,0].set_xlabel("kurto")
axs[1,0].set_ylabel("freq")
axs[1,0].set_title("hist of kurto")

#skew
axs[1,1].hist(skew,bins=10,alpha=0.7,color='green')
axs[1,1].set_xlabel("skew")
axs[1,1].set_ylabel("freq")
axs[1,1].set_title("hist of skew")

plt.tight_layout()
plt.show()
```
