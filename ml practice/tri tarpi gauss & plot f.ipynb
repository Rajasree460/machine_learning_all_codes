```python
# tri tarpi gauss & plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#define set A
np.random.seed(42)
# #user i/p
# a=int(input("enter the low val:"))
# b=int(input("enter the high val:"))
# A=np.arange(a,b+1)
A=np.arange(1,11)
ms_A=np.random.rand(len(A))

#define MFs
def tri(x,a,b,c):
    return np.maximum(np.minimum((x-a)/(b-a),(c-x)/(c-b)),0)

def trapi(x,a,b,c,d):
    return np.maximum(np.minimum(np.minimum((x-a)/(b-a),1),(d-x)/(d-c)),0)

def gauss(x,mu,sigma):
    return norm.pdf(x,mu,sigma)


#define x(an array of 1000 evenly spaced vals starting from 0 ending at 11)
x=np.linspace(0,11,1000)


#user i/ps & ms vals
tri_params=tuple(map(float,input("enter  the val of a, b, c seperated by spaces:").split()))

tri_ms=tri(x,*tri_params)


trapi_params=tuple(map(float,input("enter the val of a, b, c, d seperated by spaces:").split()))

trapi_ms=trapi(x,*trapi_params)


gauss_params=tuple(map(float,input("enter the val of mu & sigma seperated by spaces:").split()))

gauss_ms=gauss(x,*gauss_params)


#plotting
plt.figure(figsize=(12,8))
plt.scatter(A,ms_A,color='red',zorder=5)
plt.plot(x,tri_ms,label="triangular MF")
plt.plot(x,trapi_ms,label="tarpizoidal MF")
plt.plot(x,gauss_ms,label="gaussian MF")
plt.xlabel("vals or eles")
plt.ylabel("ms vals")
plt.title("tri trapi gauss plotting")
plt.legend()
plt.show()
```
