```python
#4 print on comp rel

import numpy as np
import matplotlib.pyplot as plt

#define R & S
np.random.seed(42)
R=np.random.rand(5,6)
S=np.random.rand(6,3)

#define composite relationships
def max_min(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max(np.minimum(R[i,:],S[:,j]))

    return res


def min_max(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.min(np.maximum(R[i,:],S[:,j]))

    return res

def max_product(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max(R[i,:]*S[:,j])

    return res

def max_avg(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))

    for i in range(R.shape[0]):
        for j in range(S.shape[1]):
            res[i,j]=np.max((R[i,:]+S[:,j])/2)

    return res


max_min_res=max_min(R,S)
min_max_res=min_max(R,S)
max_product_res=max_product(R,S)
max_avg_res=max_avg(R,S)


#print
print("R:\n",R)
print("\nS:\n",S)
print("\nmax min comp:\n",max_min_res)
print("\nmin max comp:\n",min_max_res)
print("\nmax product comp:\n",max_product_res)
print("\nmax avg comp:\n",max_avg_res)

#plotting
def plot_mat(mat,title):
    plt.figure(figsize=(12,8))
    plt.imshow(mat,cmap='viridis',aspect='auto')
    plt.colorbar()
    plt.title(title)
    plt.show()

plot_mat(max_min_res,"max min comp")
plot_mat(min_max_res,"min max comp")
plot_mat(max_product_res,"max product comp")
plot_mat(max_avg_res,"max avg comp")
```
