```python
#cog wav & plot

import numpy as np
import matplotlib.pyplot as plt

#define A & B
np.random.seed(42)
x=np.random.randint(1,101,20) #A==x==B

#define ms A & B
ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#define cog & wav
cog_A=np.sum(x*ms_A)/np.sum(ms_A)
cog_B=np.sum(x*ms_B)/np.sum(ms_B)
wav_A=np.sum(x*ms_A)/np.sum(ms_A)
wav_B=np.sum(x*ms_B)/np.sum(ms_B)

#print
print("A & B:",x)
print("\nms of A:",ms_A)
print("\nms of B:",ms_B)
print("\ncog of A:",cog_A)
print("\ncog of B:",cog_B)
print("\nwav of A:",wav_A)
print("\nwav of B:",wav_B)

#plot
#A
plt.figure(figsize=(12,8))
plt.scatter(x,ms_A,color='orange',label="membership vals")
plt.plot(x,ms_A,color='orange')
plt.axvline(cog_A,color='yellow',linestyle='--',label=f'cog={cog_A:.2f}')
plt.axvline(wav_A,color='blue',linestyle=':',label=f'wav={wav_A:.2f}')
plt.title("fuzzy set A")
plt.xlabel("vals or A or X or eles")
plt.ylabel("membership vals")
plt.legend()
plt.show()

#B
plt.figure(figsize=(12,8))
plt.scatter(x,ms_B,color='orange',label="membership vals")
plt.plot(x,ms_B,color='orange')
plt.axvline(cog_B,color='yellow',linestyle='--',label=f'cog={cog_B:.2f}')
plt.axvline(wav_B,color='blue',linestyle=':',label=f'wav={wav_B:.2f}')
plt.title("fuzzy set B")
plt.xlabel("vals or B or X or eles")
plt.ylabel("membership vals")
plt.legend()
plt.show()
```
