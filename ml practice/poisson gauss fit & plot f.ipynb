```python
#poisson gauss fit & plot 

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#define the random vector
np.random.seed(42)
n=150
random=np.random.randint(1,101,n)

#poisson dist
lambda_poisson=np.mean(random)

#gaussian dist
mu,std=np.mean(random),np.std(random)

#define x(an array of 150 eles)
x=np.arange(0,150)

#define poisson & gaussian fit
poisson_pmf=stats.poisson.pmf(x,lambda_poisson)
gaussian_pdf=stats.norm.pdf(x,mu,std)


#plotting
plt.figure(figsize=(12,8))
plt.hist(random,bins=20,alpha=0.6,density=True,color='g',label="data")
plt.plot(x,poisson_pmf,'bo',ms=9,label=f'poisson fit:{lambda_poisson:.2f}')
plt.plot(x,gaussian_pdf,'r-',lw=3,label=f'mu:{mu:.2f} & sigma:{std:.2f}')
plt.xlabel("vals")
plt.ylabel("probability distribution")
plt.title("fitting & plottig poisson & gaussian dist")
plt.legend(loc='best')
plt.show()
```
