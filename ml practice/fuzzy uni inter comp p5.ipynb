```python
#fuzzy uni inter comp

import numpy as np
import matplotlib.pyplot as plt

#define A & B
np.random.seed(42)
x=np.arange(100,201)

#define ms vals
ms_A=np.random.rand(len(x))
ms_B=np.random.rand(len(x))

#uni
fuzzy_union=np.maximum(ms_A,ms_B)
#inter
fuzzy_inter=np.minimum(ms_A,ms_B)
#comp
comp_A=np.subtract(1,ms_A)
comp_B=np.subtract(1,ms_B)

#plotting
plt.figure(figsize=(12,8))
plt.plot(x,ms_A,label="A",marker='o')
plt.plot(x,ms_B,label="B",marker='o')
plt.plot(x,fuzzy_union,label="A U B",marker='o')
plt.plot(x,fuzzy_inter,label="A  B",marker='o')
plt.plot(x,comp_A,label="A comp",marker='o')
plt.plot(x,comp_B,label="B comp",marker='o')
plt.xlabel("vals")
plt.ylabel("ms vals")
plt.legend()
plt.grid(True)
plt.show()
```
