```python
#mat stats mvks

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

#define matrix
np.random.seed(42)
mat=np.random.rand(100,10)

#convert matrix into dataframe
data=pd.DataFrame(mat)

#calc
mean=data.mean()
varience=data.var()
kurtosis=stats.kurtosis(data)
skewness=stats.skew(data)

#print
print("mean:\n",mean)
print("\nvarience:\n",varience)
print("\nkurtosis:\n",kurtosis)
print("\nskewness:\n",skewness)
print("\noverall mean",np.mean(mean))
print("\noverall varience:",np.var(varience))
print("\noverall kurtosis:",stats.kurtosis(kurtosis))
print("\noverall skewness:",stats.skew(skewness))

#plotting
plt.figure(figsize=(12,8))
plt.plot(mean,label="mean",marker='o')
plt.plot(varience,label="varience",marker="o")
plt.plot(kurtosis,label="kurtosis",marker='s')
plt.plot(skewness,label="skewness",marker='x')
plt.legend()
plt.grid(True)
plt.show()

#stat prop
fig,axs=plt.subplots(2,2,figsize=(12,8))

axs[0,0].hist(mean,bins=10,alpha=.7)
axs[0,0].set_xlabel("mean")
axs[0,0].set_ylabel("freq")
axs[0,0].set_title("maen's histogram")

axs[0,1].hist(varience,bins=10,alpha=.7)
axs[0,1].set_xlabel("var")
axs[0,1].set_ylabel("freq")
axs[0,1].set_title("var's histogram")

axs[1,0].hist(kurtosis,bins=10,alpha=.7)
axs[1,0].set_xlabel("kurtosis")
axs[1,0].set_ylabel("freq")
axs[1,0].set_title("kurtosis's histogram")

axs[1,1].hist(skewness,bins=10,alpha=.7)
axs[1,1].set_xlabel("skewness")
axs[1,1].set_ylabel("freq")
axs[1,1].set_title("skewness's histogram")

plt.show()
```
