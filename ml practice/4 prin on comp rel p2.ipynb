```python
#4 prin on comp rel

import numpy as np
import matplotlib.pyplot as plt

#define R & S
np.random.seed(42)
R=np.random.rand(5,6)
S=np.random.rand(6,3)

#define 4 prins on R comp S
def max_min(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))
    for i in range(R.shape[0]):
        for j in  range(S.shape[1]):
            res[i,j]=np.max(np.minimum(R[i,:],S[:,j]))

    return res 

def min_max(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))
    for i in range(R.shape[0]):
        for j in  range(S.shape[1]):
            res[i,j]=np.min(np.maximum(R[i,:],S[:,j]))

    return res 

def max_product(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))
    for i in range(R.shape[0]):
        for j in  range(S.shape[1]):
            res[i,j]=np.max(R[i,:]*S[:,j])

    return res 

def max_avg(R,S):
    res=np.zeros((R.shape[0],S.shape[1]))
    for i in range(R.shape[0]):
        for j in  range(S.shape[1]):
            res[i,j]=np.max((R[i,:]+S[:,j])//22)

    return res 

#calc 4 prins on comp rel
max_min_res=max_min(R,S)
min_max_res=min_max(R,S)
max_product_res=max_product(R,S)
max_avg_res=max_avg(R,S)

#print
print("R:\n",R)
print("S:\n",S)
print("max-min composition:\n",max_min_res)
print("min-max composition:\n",min_max_res)
print("max-product composition:\n",max_product_res)
print("max-avg composition:\n",max_avg_res)


#plotting
def plot_mat(matrix,title):
    plt.figure(figsize=(12,8))
    plt.imshow(matrix,cmap='viridis',aspect='auto') #rep matrix as img by using viridis color map
    plt.title(title)
    plt.colorbar()
    plt.show()

plot_mat(max_min_res,"max-min composition")
plot_mat(min_max_res,"min-max composition")
plot_mat(max_product_res,"max-product composition")
plot_mat(max_avg_res,"max-avg composition")
```
