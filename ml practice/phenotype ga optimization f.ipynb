```python
#phenotype ga optimization 

import random 

#define fitness func
def fitness_func(chromo):
    a,b,c,d,e,f,g,h=map(int,list(chromo))
    return (a+b)-(c+d)+(e+f)-(g+h)

#define crossovers
def onepoint_crossover_mid(par1,par2):
    midpoint=len(par1)//2
    
    child1=par1[:midpoint]+par2[midpoint:]
    child2=par2[:midpoint]+par1[midpoint:]

    return child1,child2

def onepoint_crossover_b_f(par1,par2):
    pos_b=2
    pos_f=6

    child1=par1[:pos_b-1]+par2[pos_b-1]+par1[pos_b:pos_f-1]+par2[pos_f-1]+par1[pos_f:]
    child2=par2[:pos_b-1]+par1[pos_b-1]+par2[pos_b:pos_f-1]+par1[pos_f-1]+par2[pos_f:]

    return child1,child2

def uniform_crossover(par1,par2):
    child1=''
    child2=''

    for i in range(len(par1)):
        if random.random()>0.5:
            child1+=par1[i]
            child2+=par2[i]
        else:
            child1+=par2[i]
            child2+=par1[i]

    return child1,child2


#calc fitness scores diff in b/w 2 highest rank chromosome
def calc_diff(fitness_scores):
    sorted_scores=sorted(fitness_scores,reverse=True)
    return sorted_scores[0]-sorted_scores[1]


#main func
def phenotype_ga():
    population=['65413532', '87126601', '23921285', '41852094']
    fitness_scores=[fitness_func(chromo) for chromo in population]
    ranked_population=sorted(zip(population,fitness_scores),key=lambda x:x[1],reverse=True)
    initial_diff=calc_diff(fitness_scores)

    #print initial population & fitness
    print("initial population & fitness")
    for chromo,fitness in ranked_population:  #zip(population,fitness_scores)
        print(f'chromosome:{chromo} & fitness scores:{fitness}')

    #crossovers
    child1,child2=onepoint_crossover_mid(ranked_population[0][0],ranked_population[1][0])

    child3,child4=onepoint_crossover_b_f(ranked_population[1][0],ranked_population[2][0])

    child5,child6=uniform_crossover(ranked_population[0][0],ranked_population[2][0])

    #new population,fitness & related infos
    new_population=[child1,child2,child3,child4,child5,child6]
    new_fitness_scores=[fitness_func(chromo) for chromo in new_population]
    new_ranked_population=sorted(zip(new_population,new_fitness_scores),key=lambda x:x[1],reverse=True)
    new_diff=calc_diff(new_fitness_scores)

    #print new population &fitness
    print("\nnew population & fitness")
    for chromo,fitness in new_ranked_population:  #zip(new_population,new_fitness_scores)
        print(f'chromosome:{chromo} & fitness scores:{fitness}')

    if new_diff<initial_diff:
        print(f'\noptimised chromosome:{new_ranked_population[0][0]} & {new_ranked_population[1][0]} with fitness scores:{new_ranked_population[0][1]} &{new_ranked_population[1][1]}')
    else:
        print(f'\noptimised chromosome:{ranked_population[0][0]} & {ranked_population[1][0]} with fitness scores:{ranked_population[0][1]} &{ranked_population[1][1]}')


#func call
phenotype_ga()        
```
